1. **Difference Between Simple and Multiple Linear Regression**:
   - **Simple Linear Regression**: Models the relationship between one predictor variable and one outcome variable.
   - **Multiple Linear Regression**: Models the relationship between multiple predictor variables and one outcome variable. It can also model the relationship for each category of a categorical variable if indicator variables are used.
   - The key difference is that **multiple linear regression** allows for more complex relationships and comparisons across groups within the data, making it useful for analyzing the effects of multiple variables and group differences.

2. **Indicator (Dummy) Variables in Simple Linear Regression**:
   - **Continuous Variable**: Represents the strength and direction of the relationship between predictor and outcome variables across the entire dataset.
   - **Indicator (Dummy) Variable**: Represents differences between categories within the dataset, with one category often treated as the baseline.
   - For categorical variables, dummy coding is used to create binary variables (0 or 1) to represent each category (except the baseline category).
   - **Example**: The model for a categorical variable like education level might look like:
     \[
     \text{Outcome} = \beta_0 + \beta_1 \cdot 1[\text{Bachelor's}] + \beta_2 \cdot 1[\text{Master's}] + \epsilon
     \]

3. **Effect of Adding an Interaction Between Continuous and Indicator Variables in Multiple Linear Regression**:
   - **Interaction** between a continuous and an indicator variable allows the model to capture how the relationship between the predictor and outcome varies across different categories.
   - **Example**: If studying the effect of **Age (continuous variable)** and **Gender (indicator variable)** on salary, adding an interaction term allows the model to estimate different slopes for each gender group.
   - The interaction term in the model would look like: 
     \[
     \text{Outcome} = \beta_0 + \beta_1 \cdot \text{Age} + \beta_2 \cdot 1[\text{Gender}] + \beta_3 \cdot (\text{Age} \times 1[\text{Gender}]) + \epsilon
     \]
   - This allows for differences in the effect of **Age** on **Salary** based on **Gender**.

4. **Behavior of Multiple Linear Regression Using Only Indicator Variables**:
   - When a categorical variable with more than two categories is used, it’s encoded into binary variables (indicator variables) to represent each category, with one category used as the baseline.
   - The model will estimate the effect of each category compared to the baseline, and this allows for **comparison across groups** within the data.
   - **Example**: For an education variable with three categories (High School, Bachelor's, Master's), the baseline would be High School, and the model would estimate how **Bachelor’s** and **Master’s** differ from **High School**.

5. **Relationship Between Binary and Categorical Variables**:
   - **Categorical Variables**: Variables that have more than two categories (e.g., color, education level).
   - **Binary Variables**: A type of categorical variable with only two categories, often represented as 0 and 1.
   - **Dummy Encoding**: A technique to convert non-binary categorical variables (with more than two categories) into binary variables (0 or 1) for regression analysis.
   - This encoding helps in performing regression analyses where categories are compared to the baseline.

6. **General Use of Indicator Variables in Regression**:
   - Indicator variables allow categorical data to be incorporated into regression models, providing a way to estimate the effect of each category.
   - The inclusion of indicator variables allows for comparisons of how different categories influence the outcome variable while keeping other variables constant.

   \text{Outcome} = \beta_0 + \beta_1 \times \text{Online} + \beta_2 \times \text{TV} + \beta_3 \times (\text{Online} \times \text{TV})

````python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px
import plotly.graph_objects as go
from sklearn.model_selection import train_test_split

# Assuming 'pokeaman' is your Pokémon dataset (replace with actual dataset)
# Clean up the dataset by replacing NaN values in the 'Type 2' column with 'None'
pokeaman.fillna('None', inplace=True)

# Set the number of repetitions for the simulation
reps = 100

# Arrays to store R-squared values
in_sample_Rsquared = np.array([0.0] * reps)
out_of_sample_Rsquared = np.array([0.0] * reps)

fifty_fifty_split_size = int(pokeaman.shape[0]*0.5)

# Run the simulation multiple times
for i in range(reps):
    # Split the data into training and testing sets
    pokeaman_train,pokeaman_test = train_test_split(pokeaman, train_size=fifty_fifty_split_size)

    # Define the linear model (formula: HP ~ Attack + Defense)
    model_spec = smf.ols(formula='HP ~ Attack + Defense', data=pokeaman_train)
    
    # Fit the model
    model_fit = model_spec.fit()
    
    # In-sample R-squared
    in_sample_Rsquared[i] = model_fit.rsquared

    # Out-of-sample R-squared
    yhat_model = model_fit.predict(pokeaman_test)
    y = pokeaman_test.HP
    out_of_sample_Rsquared[i] = np.corrcoef(y, yhat_model)[0, 1] ** 2

# Store results in a DataFrame
df = pd.DataFrame({
    "In Sample Performance (Rsquared)": in_sample_Rsquared,
    "Out of Sample Performance (Rsquared)": out_of_sample_Rsquared
})

# Create scatter plot to visualize the relationship between in-sample and out-of-sample performance
fig = px.scatter(df, x="In Sample Performance (Rsquared)", 
                     y="Out of Sample Performance (Rsquared)", 
                     title="In-Sample vs. Out-of-Sample R-squared Performance")
fig.add_trace(go.Scatter(x=[0, 1], y=[0, 1], name="y=x", line_shape='linear'))  # Add y=x line for reference
fig.show(renderer = 'png')
````
